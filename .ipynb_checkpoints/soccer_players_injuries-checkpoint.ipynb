```python
# !pip install japanize-matplotlib

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
import japanize_matplotlib

import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
plt.style.use('bmh')
```

```python
tuples = []
i=0
for t in open('Final-player.txt'):
    i=i+1
    if i == 2580:
        tuples.append(eval(t[1:-1]))
    else:
        tuples.append(eval(t[1:-2]))
df = pd.DataFrame(tuples)
df.columns = ["id", "name", "club","club_value","birth","weight","height","country","role","foot","transfers","injuries"]
df.head(10)
```

### Get only the name of the injury

```python
# df['injuries'].sum()
injuries_list = [injurie[1] for injurie in df['injuries'].sum()]
i = pd.Series(injuries_list).value_counts()
# (i >50).index
for injur in (i[i > 50].index):
    print(injur)
```
